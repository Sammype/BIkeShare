One quick thing to note is that there is not that much for me to do during this process as many of the steps that could be done were already completed on the inital dataset or were completed when I was working on my wrangling segment as I did them early. So besides making sure that the features I have are what I need I will basically be using this part of the project to sperate my dataset into train and test splits, so this shouldn't be all too long.

```python
#First Run Imports (I only really need pandas and the train test split, but I'll get numpy just in case)
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
#And now set the cleaned dataset to df so that I can make the changes to it.
df = pd.read_csv('data/bikes_prepped.csv')
#And because I forgot to not save the index col, drop it
df = df.iloc[: , 1:]
df.head()
```

```python
#A quick check of the columns
print(df.columns)
```

```python
#As the purpose of this project is to have an model look at the weather factors and change the price according to that, I'll
#go ahead and drop the date column now.
df.drop(columns = 'date', inplace = True)
```

```python
df.head()
```

```python
#And things are already looking pretty good for me. I have 4 features that are normalized on a scale from 0-1 (they were 
#initally) and 37 categorical features. Now I just need to seperate this out into X and y features and then seperate into 
#training and testing sets.
X = df.drop(columns = 'cnt', axis = 1)
y = df[['cnt']]
#I'll do a 25-75% split and my usual random state I go to for repeatability is 42.
X_tr, X_te, y_tr, y_te = train_test_split(X, y, test_size = 0.25, random_state = 42)
```

```python
print(len(X_tr), len(X_te))
print(len(y_tr), len(y_te))
```

```python
#I may reduce test size if needed later, but I feel like thats a pretty good split as it stands. Now lets take a peek at
#the training rows to make sure I didn't mess anything up.
X_tr
```

```python
#Alright, everything looks ready for modeling. Like I said, most of what was needed done had already been done. Now time 
#to export my training and testing files for the modeling (without forgetting the index = False argument this time).
X_tr.to_csv('data/X_train.csv', index = False)
X_te.to_csv('data/X_test.csv', index = False)
y_tr.to_csv('data/y_train.csv', index = False)
y_te.to_csv('data/y_test.csv', index = False)
```
