```python
#Run imports
import pandas as pd
import numpy as np
#I probably won't do a whole lot of visualizations but I'll import these just in case.
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
```

```python
#And take in my dataset from the data folder as a dataframe.
df = pd.read_csv('data/hour.csv')
df.head()
```

```python
#For this project I will be working with the hour by hour data as then I can identify hours where there is higher ridership
#and lower ridership as that will be benefical to the varialbe pricing model.
df.describe()
```

```python
df.info()
```

```python
#It looks like there are no null values, but that does not mean there is nothing to be done. First, I'll change the dteday
#column to a datetime object(for a full description of what each column is go to 
#(https://archive.ics.uci.edu/ml/datasets/bike+sharing+dataset)
df['date'] = pd.to_datetime(df['dteday'].astype(str), infer_datetime_format=True)
df['date']
```

```python
#And now to drop the features I don't need (dteday as I just made the replacement for it and year and month as they just
#repeat information from the date column as a category row which will eventially create additional dimensionality and 
#instant because it is just another column that basically matches the index +1 and I don't think that it is needed).
df.drop(columns = ['dteday', 'yr', 'mnth', 'instant'], inplace = True)
df.info()
```

```python
df.head()
```

```python
#Okay, now that we have gotten rid of some of the unneeded columns, it's time to expand the categorical variables into their
#own features so that regression can be done later (this doesn't need to be done right now, but I don't have many other 
#transformations in mind right now that need done so I'm just going to go with this.)

#I want the dummy rows to actually describe what each thing is, so I'll reassingn the row values to what they are before I 
#get the dummy rows
w_days = ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']
w_sit = ['Best', 'Okay', 'Bad', 'Worst']
season_d = ['Winter', 'Spring', 'Summer', 'Fall']
#Hours are already fine
```

```python
new_season = []
for item in df['season']:
    if item == 1:
        new_season.append(season_d[0])
    elif item == 2:
        new_season.append(season_d[1])
    elif item == 3:
        new_season.append(season_d[2])
    elif item == 4:
        new_season.append(season_d[3])
    else:
        raise Exception("Number outside of expectation recieved.")
```

```python
new_days = []
for item in df['weekday']:
    if item == 0:
        new_days.append(w_days[0])
    elif item == 1:
        new_days.append(w_days[1])
    elif item == 2:
        new_days.append(w_days[2])
    elif item == 3:
        new_days.append(w_days[3])
    elif item == 4:
        new_days.append(w_days[4])
    elif item == 5:
        new_days.append(w_days[5])
    elif item == 6:
        new_days.append(w_days[6])
    else:
        raise Exception("Number outside of expectation recieved.")
```

```python
#And finally the weather situation
new_wsit = []
for item in df['weathersit']:
    if item == 1:
        new_wsit.append(w_sit[0])
    elif item == 2:
        new_wsit.append(w_sit[1])
    elif item == 3:
        new_wsit.append(w_sit[2])
    elif item == 4:
        new_wsit.append(w_sit[3])
    else:
        raise Exception("Number outside of expectation recieved.")
```

```python
#Okay, now time to replace the old columns with the new ones.
df['weathersit'] = new_wsit
df['weekday'] = new_days
df['season'] = new_season
```

```python
#And now I'm ready to make the descriptive dummy rows
df_final = pd.get_dummies(df, columns = ['season', 'hr', 'weekday', 'weathersit'], drop_first = True)
print(df_final.columns)
```

```python
#As I'm making the df_final under the idea of it being mostly ready for modeling I also want to get rid of the casual and
#registered counts as I only care about the total for the regression. 
df_final.drop(columns = ['casual', 'registered'], inplace = True)
df_final.head()
```
